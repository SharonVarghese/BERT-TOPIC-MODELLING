```python
with open("posts.txt", "r") as outfile:
  docArray = [line.strip() for line in outfile]
```

```python
docArray
```

```python
from bertopic import BERTopic

topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topics, probs = topic_model.fit_transform(docArray)
```

```python
freq = topic_model.get_topic_info(); freq.head(5)
```

```python
topic_model.get_topic(0)
```

```python
topic_model.visualize_hierarchy(top_n_topics=4)
```

```python
topic_model.visualize_barchart(top_n_topics=4)
```

```python
topic_model.visualize_distribution(probs[10], min_probability=0.015)
```

```python
topic_model.visualize_heatmap(n_clusters=2, width=500, height=500)
```

```python
topic_model.update_topics(docArray, topics, n_gram_range=(1, 2))
```
